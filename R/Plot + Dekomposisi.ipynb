# Plot

```python
import pandas as pd
from pandas import read_excel
import matplotlib.pyplot as plt
series = pd.read_excel('audio_features.xlsx')
series["Date"] = pd.to_datetime(series["Date"])
```

```python
date = series["Date"]
danceability = series["danceability"]
duration = series["duration_ms"]
tempo = series["tempo"]
valence = series["valence"]
```

```python
import matplotlib.pyplot as plt
y=[danceability,duration,tempo,valence]
title=['Danceability','Duration','Tempo','Valence']
for i in range(len(y)):
    plt.figure(figsize = (15,4))
    plt.grid(True)
    plt.xlabel("Time", fontsize = 18)
    plt.xticks(fontsize=16)
    plt.yticks(fontsize=16)
    plt.plot(date,y[i])
    plt.title(title[i], fontsize = 24)
    
    # Show/save figure as desired.
    plt.show()
# Can show all four figures at once by calling plt.show() here, outside the loop.
#plt.show()
```

# Dekomposisi

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from pandas.plotting import register_matplotlib_converters
from statsmodels.tsa.seasonal import STL
from statsmodels.tsa.seasonal import seasonal_decompose

register_matplotlib_converters()
sns.set_style("darkgrid")
y=[danceability,duration,tempo,valence]
title=['Danceability','Duration','Tempo','Valence']
```

```python
dance_df=pd.DataFrame({'Date':series["Date"],'Danceability':danceability},columns=['Date','Danceability'])
dance_df
```

```python
dance_df1 = dance_df.set_index('Date')
dance_df1
```

```python
duration_df=pd.DataFrame({'Date':series["Date"],'Duration':duration},columns=['Date','Duration'])
duration_df1 = duration_df.set_index('Date')
duration_df1
```

```python
tempo_df=pd.DataFrame({'Date':series["Date"],'Tempo':tempo},columns=['Date','Tempo'])
tempo_df1 = tempo_df.set_index('Date')
tempo_df1
```

```python
valence_df=pd.DataFrame({'Date':series["Date"],'Valence':valence},columns=['Date','Valence'])
valence_df1 = valence_df.set_index('Date')
valence_df1
```

```python
danceability_additive=seasonal_decompose(dance_df1, model='additive', period=13)
fig1=danceability_additive.plot()
fig1.set_size_inches((10, 6))
plt.xlabel('Danceability')
plt.savefig('Danceability_klasik.png')
plt.show()
```

```python
duration_additive=seasonal_decompose(duration_df1, model='additive', period=13)
fig2=duration_additive.plot()
fig2.set_size_inches((10, 6))
plt.xlabel('duration')
plt.savefig('Duration_klasik.png')
plt.show()
```

```python
tempo_additive=seasonal_decompose(tempo_df1, model='additive', period=13)
fig3=tempo_additive.plot()
fig3.set_size_inches((10, 6))
plt.xlabel('Tempo')
plt.savefig('Tempo_klasik.png')
plt.show()
```

```python
valence_additive=seasonal_decompose(valence_df1, model='additive', period=13)
fig4=valence_additive.plot()
fig4.set_size_inches((10, 6))
plt.xlabel('Valence')
plt.savefig('Valence_klasik.png')
plt.show()
```

```python
danceability_STL = STL(dance_df1, period=13).fit()
fig5=danceability_STL.plot()
fig5.set_size_inches((10, 6))
plt.xlabel('Danceability')
plt.savefig('Danceability_STL.png')
plt.show()
```

```python
duration_STL = STL(duration_df1, period=13).fit()
fig6=duration_STL.plot()
fig6.set_size_inches((10, 6))
plt.xlabel('Duration')
plt.savefig('Duration_STL.png')
plt.show()
```

```python
tempo_STL = STL(tempo_df1, period=13).fit()
fig7=tempo_STL.plot()
fig7.set_size_inches((10, 6))
plt.xlabel('Tempo')
plt.savefig('Tempo_STL.png')
plt.show()
```

```python
valence_STL = STL(valence, period=13).fit()
fig8=valence_STL.plot()
fig8.set_size_inches((10, 6))
plt.xlabel('Valence')
plt.savefig('Valence_STL.png')
plt.show()
```

```python
danceability_STL1 = STL(danceability, period=13).fit()
duration_STL1 = STL(duration, period=13).fit()
tempo_STL1 = STL(tempo, period=13).fit()
valence_STL1 = STL(valence, period=13).fit()
```

```python
danceability_trend= danceability_STL1.trend.dropna().to_frame().T
duration_trend= duration_STL1.trend.dropna().to_frame().T
tempo_trend= tempo_STL1.trend.dropna().to_frame().T
valence_trend= valence_STL1.trend.dropna().to_frame().T
```

```python
trends = (danceability_trend,duration_trend,tempo_trend,valence_trend)
df=pd.concat(trends)
df
```

```python
df.to_csv('trend_components_unindexed.csv',index=False)
```
