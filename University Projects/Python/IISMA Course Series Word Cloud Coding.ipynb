```python
import numpy as np
import pandas as pd
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
from PIL import Image # to import an image: pillow as PIL
import requests
import re
```

```python
df = pd.read_csv(r'tripadvisor_hotel_reviews.csv', nrows=3000)
df.head()
```

```python
df = df.rename(columns=str.lower)
df
```

```python
text=df.review[0]
text
```

```python
wordcloud =WordCloud().generate(text)
```

```python
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis('on') #cause the default is on
plt.show
```

```python
upvote_pic = np.array(Image.open('upvote.png'))
```

```python
wc = WordCloud(width = 1960, height = 1080, min_font_size=30, max_font_size=500, 
               background_color='white', colormap= 'summer', mask = upvote_pic)
```

```python
plt.imshow(wc.generate(text), interpolation = 'bilinear')
plt.axis('off') #cause the default is on
plt.show()
```
